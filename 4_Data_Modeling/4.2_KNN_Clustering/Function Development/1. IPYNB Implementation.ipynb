```python
import re

def update_label_years(label_list, year):
    """
    Given a list of label strings and a target year, replace all year references to match
    the target year (single year and academic year formats).
    
    Args:
        label_list (list of str): Original labels with year references.
        year (int): Target year (e.g. 2023, 2022).
        
    Returns:
        list of str: Updated labels with years shifted accordingly.
    """
    updated_labels = []
    for label in label_list:
        # Replace "District YYYY" with "District {year}"
        label = re.sub(r'District \d{4}', f'District {year}', label)

        # Replace "District YYYY-YY" with "District {year-1}-{str(year)[2:]}"
        academic_year = f"{year - 1}-{str(year)[2:]}"
        label = re.sub(r'District \d{4}-\d{2}', f'District {academic_year}', label)

        updated_labels.append(label)
    return updated_labels

update_label_years(student_teacher_ratio, 2022)
update_label_years(race_ethnicity_percent, 2022)


distprof_2022_clean = pd.read_csv(r"C:\Users\mmath\OneDrive\Desktop\Capstone\HERC_Sp25\0_Datasets\1.5Data2022\District\clean_data\distprof_2022_clean.csv")
#distprof_2022_clean[update_label_years(student_teacher_ratio, 2022)]
distprof_2022_clean[update_label_years(student_teacher_ratio, 2022)]
update_label_years(staff_count, 2022)
```
